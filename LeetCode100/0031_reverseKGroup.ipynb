# 31. K 个一组翻转链表 

给你链表的头节点 `head` ，每 `k` 个节点一组进行翻转，请你返回修改后的链表。

`k` 是一个正整数，它的值小于或等于链表的长度。如果节点总数不是 `k` 的整数倍，那么请将最后剩余的节点保持原有顺序。

你不能只是单纯的改变节点内部的值，而是需要实际进行节点交换。

**示例 1：** 

![](https://assets.leetcode.com/uploads/2020/10/03/reverse_ex1.jpg)
> **输入：** head = \[1,2,3,4,5], k = 2
> **输出：** \[2,1,4,3,5]
> 

**示例 2：** 

![](https://assets.leetcode.com/uploads/2020/10/03/reverse_ex2.jpg)

> **输入：** head = \[1,2,3,4,5], k = 3
> **输出：** \[3,2,1,4,5]
> 

**提示：** 

*   链表中的节点数目为 `n`
*   `1 <= k <= n <= 5000`
*   `0 <= Node.val <= 1000`

**进阶：** 你可以设计一个只用 `O(1)` 额外内存空间的算法解决此问题吗？

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseKGroup(self, head, k):
        """
        :type head: Optional[ListNode]
        :type k: int
        :rtype: Optional[ListNode]
        """
        # 将链表按k个一组进行翻转，以实际的节点交换的形式
        # k < 链表长度
        if k == 1:
            return head
        dummy = ListNode(0, head)
        crt = dummy
        # 链表节点数
        n = 0
        while crt.next:
            crt = crt.next
            n += 1
        # 可以使用一个while循环来遍历链表
        # 接着在内部进行for _ in range(k)的循环来做一个倒转k节点
        first = dummy
        crt = head
        pre = None
        while n >= k:
            n -= k
            # 内部节点交换
            # 迭代法翻转子链表
            # 从crt开始，构建一条新的反转链表
            # crt -> nxt : nxt -> crt
            for _ in range(k):
                nxt = crt.next
                crt.next = pre
                pre = crt
                crt = nxt
            # dummy节点nxt指向原始的头节点，现存的子链表末尾节点
            nxt = first.next
            # k节点子链表被处理完后，crt指向的是下一组的头节点
            # 这一步将原本尾节点指向None的子链表链接回到链表中
            nxt.next = crt
            # pre位置在子链表头节点
            first.next = pre
            # first指向本轮处理的子链表尾节点，即下一轮的子链表头节点
            first = nxt
        
        return dummy.next
```
