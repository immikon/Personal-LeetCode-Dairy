# 10. 接雨水 

给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。

**示例 1：** 

![](https://assets.leetcode-cn.com/aliyun-lc-upload/uploads/2018/10/22/rainwatertrap.png)

> **输入：** height = \[0,1,0,2,1,0,1,3,2,1,2,1]
> **输出：** 6
> **解释：** 上面是由数组 \[0,1,0,2,1,0,1,3,2,1,2,1] 表示的高度图，在这种情况下，可以接 6 个单位的雨水（蓝色部分表示雨水）。 
> 

**示例 2：** 

> **输入：** height = \[4,2,0,3,2,5]
> **输出：** 9
> 

**提示：** 

*   `n == height.length`
*   `1 <= n <= 2 * 10^4`
*   `0 <= height[i] <= 10^5`

```python
class Solution(object):
    def trap(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        # left = 0
        # total = 0
        # temp_vec = 0
        # for right in range(len(height)):
            # ------------------------------------
            # 如何判断一个完整的容器，并计算其容积
            # left应该置于容器左壁，right置于容器右壁
            # 每确定一次容器完整后，左壁移动到右壁位置
            # ------------------------------------
            # 容器成立的条件（已知左壁高度，遍历寻找右壁）
            # 若右壁高于左壁，必然构成完整容器
            # 将小容器容积计算到总体
            # if height[right] >= height[left]:
            #     total += temp_vec
            #     temp_vec = 0
            #     left = right
            # # 其他情况
            # # 先默认左壁低于右壁计算temp_vec
            # # 若出现右壁低于左壁的容器，则无法解决
            # 
            # 
        # 改变思路：确认最高壁位置，分别从左向右/从右向左遍历
        # 需要考虑，最高点可能存在多个
        # max_p1为最左侧最高点，max_p2为最右侧最高点
        # 将整体切分为左段、中段、右段来计算
        max_p1, max_p2, max_h = (0, 0, 0)
        for p in range(len(height)):
            if(height[p] > max_h):
                max_h = height[p]
                max_p1 = p
            if(height[p] >= max_h):
                max_h = height[p]
                max_p2 = p
        left = 0
        right = len(height) - 1 
        total = 0
        temp_vec = 0
        # 左段：向右遍历
        for left1 in range(max_p1 + 1):
            if height[left1] < height[left]:
                temp_vec += height[left] - height[left1]
            if height[left1] >= height[left]:
                total += temp_vec
                temp_vec = 0
                left = left1
        # 右段：向左遍历
        for right1 in range(right, max_p2, -1):
            if height[right1] < height[right]:
                temp_vec += height[right] - height[right1]
            if height[right1] >= height[right]:
                total += temp_vec
                temp_vec = 0
                right = right1
        # 中段
        for mid in range(max_p1, max_p2 + 1):
            if height[mid] < max_h:
                temp_vec += max_h - height[mid]
            if height[mid] >= max_h:
                total += temp_vec
                temp_vec = 0

        return total

```

**纪念**

力扣完成的第一道hard，使用了双指针的思路。

利用最高点将整体切分为三段，通过三个双指针完成计算。

还能够通过动态规划的方法完成，但本方法在复杂度上已经是最优方案之一。
