# 33. 排序链表 

给你链表的头结点 `head` ，请将其按 **升序**  排列并返回 **排序后的链表**  。

**示例 1：** 

![](https://assets.leetcode.com/uploads/2020/09/14/sort_list_1.jpg)
> **输入：** head = \[4,2,1,3]
> **输出：** \[1,2,3,4]
> 

**示例 2：** 

![](https://assets.leetcode.com/uploads/2020/09/14/sort_list_2.jpg)
> **输入：** head = \[\-1,5,3,4,0]
> **输出：** \[\-1,0,3,4,5]
> 

**示例 3：** 

> **输入：** head = \[]
> **输出：** \[]
> 

**提示：** 

*   链表中节点的数目在范围 `[0, 5 * 10^4]` 内
*   `-10^5 <= Node.val <= 10^5`

**进阶：** 你可以在 `O(n log n)` 时间复杂度和常数级空间复杂度下，对链表进行排序吗？

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def sortList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # 升序排列链表
        # 第一想法：先进行一次遍历，将节点存储在以值为角标的哈希表中
        # 不大可行的感觉
        # 第二想法：在遍历链表过程中做处理，记录一个当前最大值，将小于最大值的新节点在已有节点中遍历，放在合适的位置
        # 这一做法在worst case下有n^2的时间复杂度
        if not head:
            return head
        p0 = dummy = ListNode(0, head)
        # maxval即是链表已排序部分队尾值
        maxval = head.val
        crt = head
        while crt:
            nxt = crt.next
            # 在当前节点大于已有最大值时，记录为最大值
            if crt.val >= maxval:
                maxval = crt.val
                p0 = p0.next
            # 在当前节点小于最大值时，遍历插入
            elif crt.val < maxval:
                prev = dummy
                # 遍历到prev节点的下一个节点值大于等于当前节点的位置
                while prev.next.val < crt.val:
                    prev = prev.next
                # 插入前需要清除原节点前一个节点的next联系
                # 将它与crt的下一节点建立联系，避免死循环
                p0.next = nxt
                # 将crt插入到tmp前的位置
                crt.next = prev.next
                prev.next = crt
            crt = nxt
        
        return dummy.next
```

### N^2排序算法优化后的状态

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def sortList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # 升序排列链表
        # 在遍历链表过程中做处理，记录一个当前最大值，将小于最大值的新节点在已有节点中遍历，放在合适的位置
        # 这一做法在worst case下有n^2的时间复杂度
        if not head:
            return head
        # maxvalue即是p0(已排序队尾)的值，可以放弃maxval
        dummy = ListNode(0, head)
        lastsort = head
        crt = lastsort.next
        while crt:
            # 在当前节点大于已有最大值时，记录为最大值
            if crt.val >= lastsort.val:
                lastsort = lastsort.next
            # 在当前节点小于最大值时，遍历插入
            elif crt.val < lastsort.val:
                prev = dummy
                # 遍历到prev节点的下一个节点值大于等于当前节点的位置
                while prev.next.val < crt.val:
                    prev = prev.next
                # 插入前需要清除原节点前一个节点的next联系
                # 将它与crt的下一节点建立联系，避免死循环
                lastsort.next = crt.next
                # 将crt插入到tmp前的位置
                crt.next = prev.next
                prev.next = crt
            crt = lastsort.next
        
        return dummy.next
```

### 要达到O(NlogN)需要使用二分法

这一方法在思路上是这样达成的：

- 先使用快慢指针找到链表中点

- 对左右分别进行排序

- 进行归并

# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution(object):
    def sortList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        使用归并排序，时间复杂度 O(n log n)，空间复杂度 O(log n) (递归栈)
        """
        if not head or not head.next:
            return head
        
        # 1. 分割：使用快慢指针找到链表中点
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        # slow 是中点，我们将链表从这里断开
        mid = slow.next
        slow.next = None  # 断开链表
        
        # 2. 递归：对左右两部分分别排序
        left_sorted = self.sortList(head)
        right_sorted = self.sortList(mid)
        
        # 3. 合并：将两个有序链表合并
        return self.merge(left_sorted, right_sorted)

    def merge(self, l1, l2):
        """
        合并两个有序链表
        """
        dummy = ListNode(0)
        tail = dummy
        
        while l1 and l2:
            if l1.val < l2.val:
                tail.next = l1
                l1 = l1.next
            else:
                tail.next = l2
                l2 = l2.next
            tail = tail.next
            
        # 连接剩余的部分
        if l1:
            tail.next = l1
        if l2:
            tail.next = l2
            
        return dummy.next

### 归并排序

**注意，这个方案中重要的是递归与合并的过程**

对于流程的一个小解析：

手动执行 sortList([4, 3, 2, 1])

第1层：第一次调用 sortList([4, 3, 2, 1])

- 分割:

- - head 指向 [4, 3, 2, 1]。

- - 快慢指针找到中点，slow 停在 3。

- - 链表被切成两半：head 指向 [4, 3]，mid 指向 [2, 1]。

- 递归调用 (向下一层):

- - 代码执行 left_sorted = self.sortList(head)，也就是 sortList([4, 3])。

- - 注意： 当前第1层的函数执行暂停，等待 sortList([4, 3]) 这个调用返回结果。

第2层：调用 sortList([4, 3])

- 分割:

- - head 指向 [4, 3]。

- - 快慢指针找到中点，slow 停在 4。

- - 链表被切成两半：head 指向 [4]，mid 指向 [3]。

- 递归调用 (再向下一层):

- - 代码执行 left_sorted = self.sortList(head)，也就是 sortList([4])。

- - 当前第2层的函数执行暂停，等待 sortList([4]) 返回结果。

第3层：调用 sortList([4]) - 触底反弹

- 检查终止条件:

- - if not head or not head.next: (链表只有一个节点 [4]) -> 条件为 True。

- - 函数立即 return head，也就是返回 [4]。

- - 这是递归的“最深处”，现在要开始“返回”了。

- 回到第2层：sortList([4, 3]) 继续执行
