# 13. 最小覆盖子串 

给你一个字符串 `s` 、一个字符串 `t` 。返回 `s` 中涵盖 `t` 所有字符的最小子串。如果 `s` 中不存在涵盖 `t` 所有字符的子串，则返回空字符串 `""` 。

**注意：** 

*   对于 `t` 中重复字符，我们寻找的子字符串中该字符数量必须不少于 `t` 中该字符数量。
*   如果 `s` 中存在这样的子串，我们保证它是唯一的答案。

**示例 1：** 

> **输入：** s = "ADOBECODEBANC", t = "ABC"
> **输出：** "BANC"
> **解释：** 最小覆盖子串 "BANC" 包含来自字符串 t 的 'A'、'B' 和 'C'。
> 

**示例 2：** 

> **输入：** s = "a", t = "a"
> **输出：** "a"
> **解释：** 整个字符串 s 是最小覆盖子串。
> 

**示例 3:** 

> **输入:**  s = "a", t = "aa"
> **输出:**  ""
> **解释:**  t 中两个字符 'a' 均应包含在 s 的子串中，
> 因此没有符合条件的子字符串，返回空字符串。

**提示：** 

*   `^m == s.length`
*   `^n == t.length`
*   `1 <= m, n <= 10^5`
*   `s` 和 `t` 由英文字母组成

**进阶：** 你能设计一个在 `o(m+n)` 时间内解决此问题的算法吗？

```python
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        # 目标复杂度为O(m + n)
        # 整体构造一个滑动窗口作为外部, 带来O(n)复杂度
        # 窗口滑动的逻辑应该如: 
        # 若窗口内字符类型与数量要求未达到目标 t, 则 right 右移
        # 达到目标 t要求后, left可以移动到下一位置
        # 通过一个变量 k记录是否达到要求
        # 需要一个容器 res 记录现有最小子串
        n = len(s)
        left = 0
        res_l, res_r = -1, len(s)
        # 如何记录与判断窗口字符是否达到目标
        # 使用哈希表, 利用 defaultdict(int) 构建一个字符为角标，记录出现次数的字典
        target = defaultdict(int)
        for c in t:
            target[c] += 1
        k = len(target)
        
        for right in range(n):
            target[s[right]] -= 1
            if target[s[right]] == 0:
                # 一个字符达到要求
                k -= 1
            while k == 0:   # 窗口达到要求
                if right - left < res_r - res_l:    # 更新最小字符串
                    res_r, res_l = right, left
                c = s[left]
                if target[c] == 0:  # 
                    k += 1
                target[c] += 1
                left += 1
        if res_l < 0:
            return ""
        else:
            return s[res_l: res_r + 1]
```

defaultdict(int)在初始化时，会自动为“不存在的字符”，或说无关字符赋值为0。

e.g. 初始化时t = "ABC"，那么后续访问 if target[D] == 0，就会返回true。

但这个问题并没有影响到本算法的运行，因为它通过target[x] == 0的判断机制很好地解决了无关字符的问题。

在窗口循环中，无关字符x初始时就有target[x] == 0

在right经过它时，发生了target[x] -= 1，这使得所有无关字符在初遇后记录为负数。

在left经过它时，会先判断是否为0（不为0，则可能是需要的字符，但窗口里的数量太多了，也可能是无关字符）

若是无关字符，其在内部循环中不可能产生target[x] == 0并使得条件变量k计数+1的情况，其最多只会在结尾回归到0.
